# Pitching moment around the aerodynamic center

Let us import the necessary libraries and functions in python.

```python
%run ./resources/init.py
%matplotlib inline
database = import_database_aerodynamic_center()
```

Let us consider a wing with the following data : 

```python
b=16
c_r=2.5
c_t=1
Lambda_le=0*math.pi/180
alpha0l_r=-3*math.pi/180
alpha0l_t=-1.5*math.pi/180
Cl_alphar=0.1073
Cl_alphat=0.1056
Cmac_r=-0.080
Cmac_t=-0.100
M=0.40
```

Here it is the planform of the just defined wing:

```python
plot_planform_sweep(c_r,c_t,b, Lambda_le)
```

```python
Latex(
    r'\begin{array}{rl}'
    +  r'\text{root chord,}\, c_{\mathrm{r}}: & ' + r'{0}'.format(c_r) + r'\,\text{m}'
    +  r'\\'
    +  r'\text{tip chord,}\, c_{\mathrm{t}}: & ' + r'{0}'.format(c_t) + r'\,\text{m}'
    +  r'\\'
    +  r'\text{semispan,}\, \frac{1}{2}b: & ' + r'{0}'.format(b/2) + r'\,\text{m}'
    +  r'\\'
    +  r'\text{leading edge sweep angle,}\, \Lambda_{\mathrm{le},1}: &' 
    +    r'{0}'.format(Lambda_le*180/math.pi) + r'\,\text{deg}'
    +  r'\\'
    +  r'\text{alpha 0 lift root,} \, \alpha_{0\ell,r}: &' 
    +    r'{0}'.format(alpha0l_r*180/math.pi) + r'\,\text{deg}'
    +  r'\\'
    +  r'\text{alpha 0 lift tip ,} \, \alpha_{0\ell,t}: &' 
    +    r'{0}'.format(alpha0l_t*180/math.pi) + r'\,\text{deg}'
    +  r'\\'
    +  r'\text{Root lift gradient,} \, C_{\ell_{\alpha},r}: &' 
    +    r'{0:.3}'.format(Cl_alphar*180/math.pi) + r'\,\text{rad}^{-1}'
    +  r'\\'
    +  r'\text{tip lift gradient,} \, C_{\ell_{\alpha},t}: &' 
    +    r'{0:.3}'.format(Cl_alphat*180/math.pi) + r'\,\text{rad}^{-1}'
    +  r'\\'
    +  r'\text{Root aerodynamic center moment coefficient,} \, C_{m_{ac},r}: &' 
    +    r'{0}'.format(Cmac_r) 
    +  r'\\'
    +  r'\text{Tip aerodynamic center moment coefficient,} \, C_{m_{ac},t}: &' 
    +    r'{0}'.format(Cmac_t) 
    +  r'\\'
    +  r'\text{Flight condition,} \, M: &' 
    +    r'{0}'.format(M)
    +r'\end{array}'
)
```

This wing is characterised by its straight edges and no swept leading edge. We can do some calculations about the geometry of this wing:

```python
S_ref = (c_r + c_t)*b/2
taper_ratio = c_t/c_r
AR = b*b/S_ref
print("Taper ratio: {0}\nWing area: {1} m^2\nAspect ratio: {2}".format(taper_ratio, S_ref, AR))
```

Let us calculate: $C_{M_{ac}}$ .

The angle of the quarter-chord line shall be calculated as follows :

\begin{equation}
\text{tan}\,\Lambda_{c/4} = \text{tan}\,\Lambda_{le}\,-\,\frac{(4/4)(1-\lambda)}{\text{AR}(1+\lambda)}
\end{equation}

```python
tan_lambda=tan(Lambda_le)-(4/4)*(1-taper_ratio)/AR/(1+taper_ratio)
Lambda_c=math.atan(tan_lambda)
```

```python
Latex( r'\begin{array}{ll}'
         + r'\text{tan} \Lambda_{c/4}=\text{tan}('
         + r'{0:.3}' .format(Lambda_le)
         + r'\text{rad})- \dfrac{1\cdot(1-'
         + r'{0:.3}' .format(taper_ratio)
         + r')}{'
         + r'{0:.3}' .format(AR)
         + r'\cdot(1+'
         + r'{0:.3}' .format(taper_ratio)
         + r')}='
         + r'{0:.3}' .format(tan_lambda)
         + r'\,\Rightarrow\, \Lambda_{c/4}='
         + r'{0:.3}' .format(Lambda_c)
         + r'\,\text{rad}'
         + r'\\'
         + r'\end{array}'
     )
```

The wing assigned has a small and negative $\Lambda_\mathrm{c/4}$ angle.
Let us calculate the value of the mean aerodinamic chord :


```python
c_mean= (2/3)*c_r*(1+taper_ratio+(taper_ratio)*(taper_ratio))/(1+taper_ratio)
```

```python
Latex(
r'\begin{align*}'
    + r'\bar{c} = \frac{2}{3}c_\mathrm{r}\frac{1 + \lambda + \lambda^2}{1+\lambda}'
    + r'& =' +r'{0:.4}'.format(c_mean) +r'\,\text{m}'
    + r'\end{align*}'
    
        )
```

Let us now calculate the logitudinal distance from the leading edge of the mean aerodinamic chord to the one of the root chord.

\begin{equation}
X_{le,\bar{c}} = \frac{b}{6}\,\frac{ 1 + 2\lambda}{1+\lambda}\,\tan{\Lambda_{le}}
\end{equation}

```python
tan_lambdale = math.tan(Lambda_le)
x_lec=(b/6)*((1+2*taper_ratio)/(1+taper_ratio))*tan_lambdale
```

```python
Latex(
r'\begin{align*}'
+ r'X_\mathrm{le,\bar{c}}' + r'=' + r'{0:.4}'.format(x_lec) + r'\,\text{m}'
    +r'\end{align*}'
    )
```

The distance, computed from wing root, of the mean aerodinamic chord station shall be calculated as follows : 
\begin{equation}
Y_{\bar{c}} = \frac{b}{6} \frac{1+2\lambda}{1+\lambda}
\end{equation}

```python
y_c=(b/6)*((1+2*taper_ratio)/(1+taper_ratio))
```

```python
Latex(
r'\begin{align*}'
    +r'Y_{\bar{c}}=\frac{b}{6} \frac{1+2\lambda}{1+\lambda}='
    +r'{0:.4}' .format(y_c)
   
    +r'\,\text{m}'
        + r'\end{align*}'

    )
```

## Calculate $K_2$

We calculate $K_2$ by two-dimensional interpolation over $(\Lambda_{\mathrm{le}},\mathrm{AR})$ and successively by one-dimensional interpolation over $\lambda$.

```python
# K2
var0_K2 = database['var0_K2']
var1_K2 = database['var1_K2']
var2_K2 = database['var2_K2']
data_K2 = database['data_K2']
# bracketing lambda
j_lambda = 1
g_K2_a = interp2d(var1_K2, var2_K2, data_K2[:,:,j_lambda], kind='linear')
K2_a = g_K2_a(AR, Lambda_le*180/math.pi)
K2_a = K2_a[0]
print('Lambda_LE = {0:.3} deg, AR = {1:.3}, lambda = {2:.3}\n\t --> K_2 = {3}'.format(
        Lambda_le*180/math.pi, AR, 
        var0_K2[j_lambda], K2_a))
```

This is how a first value of $K_2$ is taken from a first family of curves:

```python
plot_interpolate_K2(database['var0_K2'], database['var1_K2'], database['var2_K2'], database['data_K2'], 
                    1, # j_lambda
                    Lambda_le*180/math.pi, # Lambda_le 
                    AR) # AR
```

```python
# bracketing lambda
j_lambda = 2
g_K2_b = interp2d(var1_K2, var2_K2, data_K2[:,:,j_lambda], kind='linear')
K2_b = g_K2_b(AR, Lambda_le*180/math.pi)
K2_b = K2_b[0]
print('Lambda_LE = {0:.3} deg, AR = {1:.3}, lambda = {2:.3}\n\t --> K_2 = {3}'.format(
        Lambda_le*180/math.pi, AR, 
        var0_K2[j_lambda], K2_b))
```

```python
# %run ./src/aerodynamic_center_wing/init.py
plot_interpolate_K2(database['var0_K2'], database['var1_K2'], database['var2_K2'], database['data_K2'], 
                    2, # j_lambda
                    Lambda_le*180/math.pi, # Lambda_le 
                    AR) # AR
```

This is how the final value of $K_2$ is calulated by one-dimensional interpolation over $\lambda$ based on the previously calculated couple of values:

```python
ja = 1; jb = 2
K2 = K2_a + (taper_ratio - var0_K2[ja])*(K2_b - K2_a)/(var0_K2[jb] - var0_K2[ja])
print('K_2 = {0}'.format(K2))
```

```python
display_workflow_K2(var0_K2[ja], var0_K2[jb], K2_a, K2_b, taper_ratio, K2)
```

Similarly, we calculate $X_{\mathrm{ac}}'/c_{\mathrm{r}}$ by two-dimensional interpolation over $(\tan\Lambda_{\mathrm{le}}/\sqrt{1 - M^2},\mathrm{AR}\tan\Lambda_{\mathrm{le}})$ and successively by one-dimensional interpolation over $\lambda$.

## Calculation of  $C_{M_{ac}}$

The location of the aerodynamic wing center identifies a line normal to the center plane of the equation $X = X_\mathrm{ac}$ that can be used to calculate the moment law of the basic load :
    \begin{equation}
     x_b(Y)=X_{ac}-\Big[Ytan\Lambda_{le}+\bar{x}_{ac,2D}(Y)c(Y)\Big]
    \end{equation}

Where $\bar{x}_{ac,2D}(Y)$ represents the  law of the distance, for each Y station, between the local aerodynamic center and the leading edge of the local chord.For the assigned wing  we have : $\bar{x}_{ac,2D}(Y)=\frac{1}{4}$ and so the above formula becomes :
    \begin{equation}
     x_b(Y)=X_{ac}-\frac{1}{4}c(Y)
    \end{equation}

```python
display(Math(r'x_b(Y) =\,' + '{0:.3}'.format(X_ac) + '\,\text{m} -  ' +'\dfrac{1}{4} \cdot ( \,' + '{0:.3}'.format(B_c) + r'\,\text{m}'))
```

The pitching moment coefficient of the wing around his aerodynamic center,$C_{M_{ac}}$, is given by the formula:

\begin{equation}
C_{M_{ac}} 
  = \frac{2}{S\bar{c}} \int_0^{b/2} 
    C_{m_{ac},2D}(Y)c^2(y) +\frac{2}{S\bar{c}} \int_0^{b/2} 
    (cC_{\ell})_b(Y) \,x_b(Y)
   \, \mathrm{d}y
\end{equation}

Where:
- $ C_{m_{ac},2D}$ is the law of the moment coefficient around the local aerodynamic center.
- $ (cC_{\ell})_b(Y)$is the product between the law of the chord and the basic lift coefficient, that is present when the airfoil is at $\alpha_{0L}$.
- $x_b$ is the distance between the point of application of the basic load and the wing aerodynamic center.

We can see that $C_{M_{ac}}$ has two contributes:

- $C_{M_{ac,a}}$
- $C_{M_{ac,b}}$

These contributes are integrated along the span and are dividided for $\bar{q}_\infty S \,\bar{c}$
We can write this law for  : 
\begin{equation}
C_{m_{ac}}(Y)=A_{C_{m_{ac}}}Y+B_{C_{m_{ac}}}
\end{equation}
in wich $A_{C_{m_{ac}}}$ and $B_{C_{m_{ac}}}$ shall be calculated as follows:

```python
A_CM=(Cmac_t-Cmac_r)/b*2
B_CM=Cmac_r
Latex(
    r'\begin{equation}'

+r'A_{C_{m_{ac}}}=\frac{C_{m_{ac},t}-C_{m_{ac},r}}{b/2}='
 +r'\frac{'   
+r'{0:.3}' .format(Cmac_t)
+r'-('
    +r'{0:.3}' .format(Cmac_r)
    +r') }'
    +r'{'
    +r'{0:.3}' .format(b/2)
    +r'} ='
    +r'{0:.3}' .format(A_CM)
    +r'\, \text{m}^{-1}'
    +r'\\'
    +r'B_{C_{m_{ac}}}=C_{m_{ac},r}='
    +r'{0:.3}' .format(B_CM)
+r'\end{equation}'
)
```

And the above formula becomes : 

```python
display(Math(r'C_{m_{ac},2D}(Y) =\,' + '{0:.3}'.format(A_CM)  + r'\,\text{m}^{-1}'+ '\,Y + ' + '{0:.3}'.format(B_CM)))
```

Before calculating the integral of $C_{M_{ac,a}}$ , let us calculate the chords law of the wing : 

```python
A_c = 2*(c_t - c_r)/b; B_c = c_r;
display(Math(r'c(y) =\,' + '{0:.3}'.format(A_c) + '\,y + ' + '{0:.3}'.format(B_c) + r'\,\text{m}'))
```

And finally, by putting these two laws in the integer we can calculate : 

```python
c_maca=integrate_c_maca(A_CM,B_CM,A_c,B_c,b,S_ref)[0]
display(Math(r'C_{M_{ac,a}}\,=\,' + '{0:.3}'.format(c_maca) ))
```

```python
c_mac_law_integral_indefinite = integrate_c_maca(A_CM,B_CM,A_c,B_c,b,S_ref)[1]
display_workflow_c_maca(S_ref, b,A_CM,B_CM, A_c, B_c, c_mac_law_integral_indefinite, c_maca)
```

For a generic pitch angle at which the wing produces a lift coefficient $C_\mathrm{L}$, the wing span loading is expressed as :

\begin{equation}
\gamma(Y)=\frac{c(Y)C_{\ell}(Y)}{2b}=C_{L\gamma_{a1}}(Y)+\gamma_{b}(Y)
\end{equation}

The load is the sum of two terms:
- $C_{L\gamma_{a1}}(Y)$ that is zero when we are at alpha 0 lift.
- $\gamma_{b}(Y)$, basic loads that is zero for not twisted(geometrically and aerodynamically) airfoil.
    

In particoular:
\begin{equation}
 \int_0^{1} 
  \gamma_b(\eta)
   \, \mathrm{d}\eta
\end{equation}

In the $C_{M_{ac,b}}$ formula, we have $ \big(cC_\ell\big)_b\big(Y\big) = 2b\gamma_b\big(Y\big)$ . This load law can be determined with numerical methods or , with an accetable aproximation, from the twist , $\epsilon_g\big(Y\big) $, the aerodinamic ,$\alpha_{0\ell}\big(Y\big)$ laws and the 0-lift angle of the wing $\alpha_{0L}$.

Let us calculate $C_{M_{ac},b}$.
The approximate expression of basic load note from theory swirling wings  is:

\begin{equation}
(cC_{\ell})_b(Y)=\frac{1}{2}c(Y)C_{\ell_{\alpha}}(Y)\Bigl\{\alpha_{0L}-\Big[\alpha_{0\ell}(Y)-\epsilon_g(Y)\Big]\Bigl\}
\end{equation}

Where $\frac{1}{2}$ explains the tridimensional effect around the mid line of the wing.

Now we have to calculate $C_{\ell_{\alpha}}$.We could say that $C_{\ell_{\alpha}}=2\pi$.Alternativelly we have to find this expression:

\begin{equation}
C_{\ell_{\alpha}}=A_{C_{\ell_{\alpha}}}Y+B_{C_{\ell_{\alpha}}}
\end{equation}

### The law $\alpha_{0\ell}(y)$ of profile zero-lift angles

The function $\alpha_{0\ell}(y)$ gives the zero-lift angle of attack of the wing profile at the generic spanwise station $y\in \big[0,\frac{1}{2}b\big]$. We assume the followinh linear form

\begin{equation}
\alpha_{0\ell}(y) = A_{\alpha} \, y + B_{\alpha}
\end{equation}

with 

\begin{equation}
A_{\alpha} = \frac{2}{b} \left( \alpha_{0\ell,\mathrm{t}} - \alpha_{0\ell,\mathrm{r}} \right)
  \;\;,\qquad 
B_{\alpha} = \alpha_{0\ell,\mathrm{r}}
\end{equation}

The following code assigns the two required values at root and tip sections and calculates the above two coefficients:

```python
A_alpha = 2*(alpha0l_t- alpha0l_r)/b; 
B_alpha = alpha0l_r;
print("A_alpha: {0} rad/m\nB_alpha: {1} rad".format(A_alpha,B_alpha))
```

and the function $\alpha_{0\ell}(y)$ becomes:

```python
display(Math(
        r'\alpha_{0\ell}(y) =\,' + '{0:.4}'.format(A_alpha) + r'\frac{\text{rad}}{\text{m}} \,y ' 
        + '{0:.4}'.format(B_alpha) + r'\,\text{rad}'))
```

We also need to calculate the twist law and the 2D lift gradient : 

```python
eps_tip = -2.5*math.pi/180
A_eps = 2*eps_tip/b; B_eps = 0;
print("A_eps: {0} rad/m\nB_eps: {1} rad".format(A_eps,B_eps))
```

```python
display(Math(r'\epsilon_{\mathrm{g}}(y) =\,' + '{0:.3}'.format(A_eps) + r'\frac{\text{rad}}{\text{m}}\,y'))
```

```python
A_clalpha = 2*(Cl_alphat*180/math.pi- Cl_alphar*180/math.pi)/b; 
B_clalpha = Cl_alphar*180/math.pi;
print("Acl_alpha: {0} rad/m\nBcl_alpha: {1} rad".format(A_clalpha,B_clalpha))
```

```python
display(Math(
        r'{C_{\ell_{\alpha}}\big( Y \big)} =\,' + '{0:.4}'.format(A_clalpha) + r'(\text{rad m}^{-1}) \,y+ ' 
        + '{0:.4}'.format(B_clalpha) + r'\,\text{rad}^{-1}'))
```
