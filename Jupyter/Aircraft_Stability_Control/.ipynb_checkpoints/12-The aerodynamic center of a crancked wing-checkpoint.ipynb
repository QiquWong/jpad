After we have calculated the aerodynamic center of a tapered wing, we now want to calculate the aerodynamic center for a crancked wing.

```python
plot_planform_cw(c_r, c_k, c_t, b_k, b, Lambda_le_1, Lambda_le_2)
```

The function $c(y)$ is the piecewise linear law that gives the chord at the generic spanwise station $y\in \big[0,\frac{1}{2}b\big]$. This has the form

\begin{equation}
c(y) = 
\begin{cases}
A_{c,1} y + B_{c,1}  & \text{if $\;0\le y \le \frac{1}{2}b_{\mathrm{k}}$} \\[0.5em]
A_{c,2} \big( y - \frac{1}{2}b_{\mathrm{k}} \big) + B_{c,2} & \text{if $\;\frac{1}{2}b_{\mathrm{k}} < y \le \frac{1}{2}b$}
\end{cases}
\end{equation}

with 

\begin{equation}
A_{c} = 
\begin{cases}
A_{c,1} = \dfrac{2}{b_{\mathrm{k}}} \left( c_{\mathrm{k}} - c_{\mathrm{r}} \right)  
  & \text{if $\;0\le y \le \frac{1}{2}b_{\mathrm{k}}$} \\[0.5em]
A_{c,2} = \dfrac{2}{b - b_{\mathrm{k}}} \left( c_{\mathrm{t}} - c_{\mathrm{k}} \right)  
  & \text{if $\;\frac{1}{2}b_{\mathrm{k}} < y \le \frac{1}{2}b$}
\end{cases}
\end{equation}

and

\begin{equation}
B_{c} = 
\begin{cases}
B_{c,1} = c_{\mathrm{r}}
  & \text{if $\;0\le y \le \frac{1}{2}b_{\mathrm{k}}$} \\[0.5em]
B_{c,2} = c_{\mathrm{k}}
  & \text{if $\;\frac{1}{2}b_{\mathrm{k}} < y \le \frac{1}{2}b$}
\end{cases}
\end{equation}

Let's calculate the coefficients involved in these equations.

```python
A_1 = 2*(c_k - c_r)/b1; B_1 = c_r;
A_2 = 2*(c_t - c_k)/(b2); B_2 = c_k;
Latex(
   r'\begin{array}{ll}'
     +r'A_{c,1} = \dfrac{2}{ b_{\mathrm{k}}} \left( c_{\mathrm{t}} - c_{\mathrm{k}} \right)='
     +r'{0:.3}' .format(A_1)
     +r'\\'
     +r'B_{c,1}=c_{r,1}='
     +r'{0:.4}' .format(c_r)
     +r'\,\text{m} '
     +r'\\'
     +r'A_{c,2} = \dfrac{2}{ b_{\mathrm{2}}} \left( c_{\mathrm{t,2}} - c_{\mathrm{r,2}} \right)='
     +r'{0:.3}' .format(A_2)
     +r'\\'
     +r'B_{c,2}=c_{r,2}='
     +r'{0}' .format(c_k)
     +r'\,\text{m} '
   +r'\end{array}'
)
```

So we have:

```python
Latex(
    r'\begin{equation}'
     + r'c(y) ='
     + r'\begin{cases}'
     + r'c_1(Y)='
     + r'{0:.3}' .format(A_1)
     + r'Y+'
     + r'{0:.3}' .format(B_1)
     + r'\,\text{m}'
     + r'& \text{if $\;0\,\text{m} \le Y \le'
     + r'{0:.3}' .format(b1/2)
     + r'$} \, \text{m}\\[0.5em]'
     + r'c_2(Y)='
     + r'{0:.3}' .format(A_2)
     + r'\big(Y-'
     + r'{0:.3}' .format(b1/2)
     + r'\, \text{m}\big)' 
     + r'& \text{if $\;'
     + r'{0:.3}' .format(b1/2)
     + r'\, \text{m}< Y \le'
     +r'{0:.3}' .format(b/2)
     + r'\,\text{m}$}'
     +r'\end{cases}'
    + r'\end{equation}'
      )
```

## Geometrical charateristics

For the inner panel we have:

```python
taper_1=c_k/c_r
display(Math(r'\lambda_{1} =\,' + '{0:.4}'.format(taper_1)))
```

```python
AR1=b1*b1/S1
display(Math(r'\text{aspect ratio:}\quad AR_\textrm{1} = '+ '{0:.4}'.format(AR1)))
```

```python
S1= (c_r + c_k)*b1/2
display(Math(r'S_1 =\,' + '{0:.5}' .format(S1) + r'\, \text{m}^2'))
```

Let us calculate the value of the inner mean aerodinamic chord :

```python
c_mean_1= (2/3)*c_r*(1+taper_1+(taper_1)*(taper_1))/(1+taper_1)
Latex(r'\begin{align*}'
        + r'\bar{c}_\mathrm{1} = \frac{2}{3}c_\mathrm{r,1}\frac{1 + \lambda_1 + \lambda_1^2}{1+\lambda_1}'
        + r'& =' +r'{0:.4}'.format(c_mean_1) +r'\,\text{m}'
       + r'\end{align*}' 
        )
```

Let us now calculate the logitudinal distance from the leading edge of the inner mean aerodinamic chord to the one of the root chord.

\begin{equation}
X_{le,\bar{c_1}} = \frac{b_{k,1}}{6}\,\frac{ 1 + 2\lambda_1}{1+\lambda_1}\,\tan{\Lambda_{le,1}}
\end{equation}

```python
x_lec_1=b1/6*(1+2*taper_1)/(1+taper_1)*tan(Lambda_le_1)
Latex(
r'\begin{align*}'
+ r'X_\mathrm{le,\bar{c}1}=\dfrac{b_{k,1}}{6}\dfrac{1+2\lambda_1}{1+\lambda_1}tan\Lambda_{le,1}' + r'=' + r'{0:.3}'.format(x_lec_1) + r'\,\text{m}'
    +r'\end{align*}'
    )
```

Let us calculate now the distance, computed from wing root, of the mean aerodinamic chord station .
\begin{equation}
Y_{\bar{c},1} = \frac{b_{k,1}}{6} \frac{1+2\lambda_1}{1+\lambda_1}
\end{equation}


This formula with the given inner panel shape data, gives :

```python
y_c=(b_k/6)*((1+2*taper_1)/(1+taper_1))
Latex(
r'\begin{align*}'
    + r'Y_\mathrm{\bar{c},1} = \frac{b_k}{6}\frac{1 + 2\lambda_1}{1+\lambda_1}'
    + r'& =' +r'{0:.4}'.format(y_c) +r'\,\text{m}' 
    + r'\end{align*}'
    
        )
```

Let us do the same for  outer panel, by using the $c_k$ instead of $c_r$ and $c_t$ instead theof $c_k$. 

\begin{equation}
\lambda_2 = \frac{ c_{\mathrm{k}} }{ c_{\mathrm{r}} }
\end{equation}

```python
taper_2=c_t/c_k
display(Math(r'\lambda_{2} =\,' + '{0:.3}'.format(taper_2)))
```

And then we can define the inner aspect ratio 
\begin{equation}
%\label{eq:AspectRatio}
AR_\mathrm{2} = \frac{ b_{2}^2 }{ S }
\end{equation}

```python
AR_2=b2*b2/S2
display(Math(r'\text{aspect ratio:}\quad AR_2 = '+ '{0:.4}'.format(AR_2)))
```

```python
S2= (c_t + c_k)*b2/2
display(Math(r'S_2 =\,' + '{0:.5}' .format(S2) + r'\, \text{m}^2'))
```

```python
c_mean_2= (2/3)*c_k*(1+taper_2+(taper_2)*(taper_2))/(1+taper_2)
Latex(
r'\begin{align*}'
    + r'\bar{c}_\mathrm{2} = \frac{2}{3}c_\mathrm{r,2}\frac{1 + \lambda_2 + \lambda_2^2}{1+\lambda_2}'
    + r'& =' +r'{0:.4}'.format(c_mean_2) +r'\,\text{m}'
    + r'\end{align*}'
    
        )
```

So the difference between the leading edge of $\bar{c_2}$ and the kink station is:

\begin{equation}
X_{le,\bar{c_2}} - X_B= \frac{b}{6}\,\frac{ 1 + 2\lambda_2}{1+\lambda_2}\,\tan{\Lambda_{le,2}}
\end{equation}

```python
tan_lambda_2 = math.tan(Lambda_le_2)
x_lec_2=b1/2*tan(Lambda_le_1)+(b2/6)*((1+2*taper_2)/(1+taper_2))*tan(Lambda_le_2)
Latex(
r'\begin{align*}'
+ r'X_\mathrm{le,\bar{c}2}' + r'=' + r'{0:.4}'.format(x_lec_2) + r'\,\text{m}'
    +r'\end{align*}'
    )
```

The actual distance from the plane of symmetry of the wing, of the outer mean aerodinamic chord station is given by :

```python
y_c_2=(b1/2) + (b2/6)*((1+2*taper_2)/(1+taper_2))
Latex(
r'\begin{align*}'
    + r'Y_\mathrm{\bar{c},2} = \frac{b_{1}}{2} + \frac{b_{2}}{6}\frac{1 + 2\lambda_2}{1+\lambda_2}'
    + r'& =' +r'{0:.4}'.format(y_c_2) +r'\,\text{m}' 
    + r'\end{align*}'
    
        )
```

```python
S=S1+S2
Latex(
r'\begin{align*}'
+ r'S &' + r'='+r'S_1\,+\,S_2' + r'\,=\,'
    + r'{0:.4}'.format(S1) + r'\,\text{m}^2' +r'\,+\,'  + r'{0:.4}'.format(S2) + r'\,\text{m}^2' + r'\,=\,'
    + r'{0:.4}'.format(S) + r'\,\text{m}^2'
    +r'\end{align*}'
)
```

```python
AR=b*b/S
display(Math(r'\text{aspect ratio:}\quad AR = '+ '{0:.4}'.format(AR)))
```

We have a simple formula to calculate $\bar{c}$ :

\begin{equation}
\bar{c} = \frac{ S_1\bar{c}_1 + S_2\bar{c}_2 } { S_1 + S_2 }
\end{equation}

```python
c_mean = (S1*c_mean_1 + S2*c_mean_2) / (S1 + S2)
display(Math(r'\text{Mean Aerodinamic Chord:}\quad \bar{c} = '+ '{0:.4}'.format(c_mean) + r'\,\text{m}'))
```

We have $\bar{c}$>c_{t,1}, so the y station of the mean aerodynamic chord has to be searched between the y of the inner panel,using the $c_1(y) $ law:

\begin{equation}
\bar{c}=A_{c,1}Y_{\bar{c}}+B_{c,1} =>  Y_{\bar{c}}=\frac{\bar{c}-B_{c,1}}{A_{c,1}}
\end{equation}

```python
y_mean=(c_mean-B_1)/A_1
Latex(
  r'\begin{align*}'
    +r'Y_{\bar{c}}=\frac{\bar{c}-B_{c,1}}{A_{c,1}}='
    +r'{0:.3}'  .format(y_mean)
    +r'\ m'
  +r'\end{align*}'
)
```

```python
tan_lambda_1 = math.tan(Lambda_le_1)
x_lec=y_mean*tan_lambda_1
Latex(
   r'\begin{align*}'
    +r'X_{le,\bar{c}}=Y_{\bar{c}}tan\Lambda_{le,1}='
    +r'{0:.3}' .format(x_lec)
    +r'\ \text{m}'
   +r'\end{align*}'
)
```

```python
plot_planform_cw(c_r, c_k, c_t, b_k, b, Lambda_le_1, Lambda_le_2, 
              mac=c_mean, X_le_mac=x_lec, Y_mac=y_mean)
```

## Functions $K_1$, $K_2$ and $X_{\mathrm{ac}}'/c_{\mathrm{r}}$

We want to calculate the aerodynamic center of this cracked wing.We have to get an extern extended panel  named-constructed outer panel- as showed in the figure.You can construct by extending the leading and trailing edge of the outer panel inward starting form the points $B$ and $B^{'}$ until you meet in the points $D$ and $D^{'}$  the straight line parallel to the x axis of equation $Y=\frac{1}{4}b_1$.So we will consider two panel:
- Inner panel: $P_1=ABB^{'}A^{'}$
- Extended panel: $P_{2^{'}}=DCC^{'}D^{'}$

```python
database = import_database_aerodynamic_center()
```

The plot representing $K_1$ as a function of taper ratio $\lambda$ is reported below.

```python
plot_K1(database['var0_K1'], database['data_K1'])
```

```python
multiplot_K2(database['var0_K2'], database['var1_K2'], database['var2_K2'], database['data_K2'])
```

## Calculate $K_1$

We calculate $K_1$ by one-dimensional interpolation.

```python
# K1
var0_K1 = database['var0_K1']
data_K1 = database['data_K1']
g_K1 = interp1d(var0_K1, data_K1)
K1 = g_K1(taper_1)
print('lambda = {0}\n\t --> K_1 = {1}'.format(taper_1,K1))
```

This is how the value of $K_1$ is taken from the curve:

```python
plot_interpolate_K1(database['var0_K1'], database['data_K1'],taper_1)
```

## Calculate $K_2$

We calculate $K_2$ by two-dimensional interpolation over $(\Lambda_{\mathrm{le}},\mathrm{AR})$ and successively by one-dimensional interpolation over $\lambda$.

```python
# K2
var0_K2 = database['var0_K2']
var1_K2 = database['var1_K2']
var2_K2 = database['var2_K2']
data_K2 = database['data_K2']
# bracketing lambda
j_lambda = 1
g_K2_a = interp2d(var1_K2, var2_K2, data_K2[:,:,j_lambda], kind='linear')
K2_a = g_K2_a(AR1, Lambda_le_1*180/math.pi)
K2_a = K2_a[0]
print('Lambda_LE = {0:.3} deg, AR = {1:.3}, lambda = {2:.3}\n\t --> K_2 = {3}'.format(
        Lambda_le_1*180/math.pi, AR1, 
        var0_K2[j_lambda], K2_a))
```

This is how a first value of $K_2$ is taken from a first family of curves:

```python
# %run ./src/aerodynamic_center_wing/init.py
plot_interpolate_K2(database['var0_K2'], database['var1_K2'], database['var2_K2'], database['data_K2'], 
                    1, # j_lambda
                    Lambda_le_1*180/math.pi, # Lambda_le 
                    AR1) # AR
```

```python
# bracketing lambda
j_lambda = 2
g_K2_b = interp2d(var1_K2, var2_K2, data_K2[:,:,j_lambda], kind='linear')
K2_b = g_K2_b(AR1, Lambda_le_1*180/math.pi)
K2_b = K2_b[0]
print('Lambda_LE = {0:.3} deg, AR = {1:.3}, lambda = {2:.3}\n\t --> K_2 = {3}'.format(
        Lambda_le_1*180/math.pi, AR1, 
        var0_K2[j_lambda], K2_b))
```

This is how a second value of $K_2$ is taken from a second family of curves:

```python
# %run ./src/aerodynamic_center_wing/init.py
plot_interpolate_K2(database['var0_K2'], database['var1_K2'], database['var2_K2'], database['data_K2'], 
                    2, # j_lambda
                    Lambda_le_1*180/math.pi, # Lambda_le 
                    AR1) # AR
```


This is how the final value of $K_2$ is calulated by one-dimensional interpolation over $\lambda$ based on the previously calculated couple of values:


```python
ja = 1; jb = 2
K2 = K2_a + (taper_1 - var0_K2[ja])*(K2_b - K2_a)/(var0_K2[jb] - var0_K2[ja])
print('K_2 = {0}'.format(K2))
```

```python
display_workflow_K2(var0_K2[ja], var0_K2[jb], K2_a, K2_b, taper_1, K2)
```

## Calculate $X_{\mathrm{ac}}'/c_{\mathrm{r}}$

Similarly, we calculate $X_{\mathrm{ac}}'/c_{\mathrm{r}}$ by two-dimensional interpolation over $(\tan\Lambda_{\mathrm{le}}/\sqrt{1 - M^2},\mathrm{AR}\tan\Lambda_{\mathrm{le}})$ and successively by one-dimensional interpolation over $\lambda$.

```python
# Xac' / c_r
var0_XacCr = database['var0_XacCr']
var1_XacCr = database['var1_XacCr']
var2_XacCr = database['var2_XacCr']
data_XacCr = database['data_XacCr']
# bracketing lambda
j_lambda = 3
g_XacCr_a = interp2d(var1_XacCr, var2_XacCr, data_XacCr[:,:,j_lambda], kind='linear')
y_ = AR1*math.tan(Lambda_le_1)
x_ = math.tan(Lambda_le_1)/math.sqrt(1 - math.pow(M,2))
XacCr_a = g_XacCr_a(y_, x_)
XacCr_a = XacCr_a[0]
print('Lambda_LE = {0:.3} deg, AR = {1:.3}, Mach = {2:.3}, lambda = {2:.3}'.format(
        Lambda_le_1*180/math.pi, AR1, M, var0_XacCr[j_lambda]))
print('x_1 = tan(Lambda_LE)/sqrt(1-M^2) = {0}\nx_2 = AR*tan(Lambda_LE) = {1}\n\t --> Xac\'/c_r = {2}'
    .format(x_, y_, XacCr_a))
```

This is how a first value of $X_{\mathrm{ac}}'/c_{\mathrm{r}}$ is taken from a first family of curves:

```python
# %run ./src/aerodynamic_center_wing/init.py
plot_interpolate_XacCr(
    database['var0_XacCr'], database['var1_XacCr'], database['var2_XacCr'], database['data_XacCr'], 
    3, # j_lambda
    Lambda_le_1*180/math.pi, # Lambda_le (deg) 
    AR1, # AR 
    M) # Mach
```

```python
# bracketing lambda
j_lambda = 4
g_XacCr_b = interp2d(var1_XacCr, var2_XacCr, data_XacCr[:,:,j_lambda], kind='linear')
y_ = AR1*math.tan(Lambda_le_1)
x_ = math.tan(Lambda_le_1)/math.sqrt(1 - math.pow(M,2))
XacCr_b = g_XacCr_b(y_, x_)
XacCr_b = XacCr_b[0]
print('Lambda_LE = {0:.3} deg, AR = {1:.3}, Mach = {2:.3}, lambda = {2:.3}'.format(
        Lambda_le_1*180/math.pi, AR1, M, var0_XacCr[j_lambda]))
print('x_1 = tan(Lambda_LE)/sqrt(1-M^2) = {0}\nx_2 = AR*tan(Lambda_LE) = {1}\n\t --> Xac\'/c_r = {2}'
    .format(x_, y_, XacCr_b))
```

This is how a second value of $X_{\mathrm{ac}}'/c_{\mathrm{r}}$ is taken from a second family of curves:

```python
# %run ./src/aerodynamic_center_wing/init.py
plot_interpolate_XacCr(
    database['var0_XacCr'], database['var1_XacCr'], database['var2_XacCr'], database['data_XacCr'], 
    4, # j_lambda
    Lambda_le_1*180/math.pi, # Lambda_le (deg) 
    AR1, # AR 
    M) # Mach
```

This is how the final value of $X_{\mathrm{ac}}'/c_{\mathrm{r}}$ is calulated by one-dimensional interpolation over $\lambda$ based on the previously calculated couple of values:

```python
ja = 3; jb = 4
XacCr = XacCr_a + (taper_1 - var0_XacCr[ja])*(XacCr_b - XacCr_a)/(var0_XacCr[jb] - var0_XacCr[ja])
print('Xac\'/c_r = {0}'.format(XacCr))
```

```python
display_workflow_XacCr(var0_XacCr[ja], var0_XacCr[jb], XacCr_a, XacCr_b, taper_1, XacCr)
```

So we obtain:

```python
x_accmean=K1*(XacCr-K2)
Latex(
     r'\begin{equation}'
        + r'\frac{x_{\mathrm{ac}}}{\bar{c}} =K_{1,P_1}\Bigl(\dfrac{X_{\mathrm{ac}}}{c_r}'
        + "'"
        + r'\Big|_{P_1}-K_{2,P1} \Bigl)=' 
        + r' {0:}' .format(K1)
        + r'\big('
        + r'{0:.3}'.format(XacCr)
        + r'- {0:.3}'.format(K2)
        + r'\big)='
        + r'{0:.3}' .format(x_accmean)
     +r'\end{equation}'
    )
```

Let us consider now the constructed outer panel:

```python
b2A=2*(b2/2+b1/4)
```

```python
Latex (
   r'\begin{equation}'
    +r'b_{2' +r"'" +r'}=2\Big(\frac{b_2}{2}+ \frac{b_1}{4} \Big)=2\big('
    +r'{0:1}' .format(b2/2)
    +r'+'
    +r'{0:.3}' .format(b1/4)
    +r'\big)='
    +r'{0:.3}' .format(b2A)
    +r'\, \text{m}'
   +r'\end{equation}'
)
```

```python
c_r2A=A_2*(-b1/4)+B_2
```

```python
Latex (
    r'\begin{equation}'
    +r'c_{r,2' +r"'" +r'}=A_{c,2}\Big(-\frac{b_1}{4} \Big)='
    +r'{0:.3}' .format(A_2)
    +r'\big(-'
    +r'{0:.3}' .format(b1/4)
    +r'\, \text{m}\big) +'
    +r'{0:.3}' .format(B_2)
    +r'\, \text{m} ='
    +r'{0:.3}' .format(c_r2A)
    +r'\, \text{m}'
    +r'\end{equation}'
)
```

Now we have to calculate the lift-gradient for both panels.

First we calculate a0:

```python
cos_Lambda_1 = math.cos(Lambda_le_1)
a0=6.088/sqrt(1-M*M*cos_Lambda_1*cos_Lambda_1)
```

```python
Latex(
    r'\begin{equation}'
    +r'a_0=\frac{6.088 \text{rad}^{-1}}{\sqrt{1-M^2cos^2(\lambda_{le,1})}}='
    +r'{0:.3}' .format(a0)
   + r'\end{equation}'
)
```

We'll calculate the coefficient by:

\begin{equation}
C_{L\alpha}\Bigr|_\textit{P1} = \frac{a_0 cos\Lambda_{le,1}}{\sqrt{ 1 - M^2 {cos}^2 \Lambda_{le,1} + \big( \frac{a_0 cos \Lambda_{le,1}}{ \pi AR_1} \big)^2 } + \frac{a_0 cos \Lambda_{le,1}}{\pi AR_1}}
\end{equation}



\begin{equation}
C_{L\alpha}\Bigr|_\textit{P2'} = \frac{a_0 cos\Lambda_{le,2}}{\sqrt{ 1 - M^2 {cos}^2 \Lambda_{le,2} + \big( \frac{a_0 cos \Lambda_{le,2}}{ \pi AR_{2'}} \big)^2 } + \frac{a_0 cos \Lambda_{le,2}}{\pi AR_{2'}}}
\end{equation}

```python
pi=3.14
```

```python
CL_alpha1=(a0*cos(Lambda_le_1))/(sqrt(1-M*M*cos(Lambda_le_1)*cos(Lambda_le_1)+a0*cos(Lambda_le_1)*a0*cos(Lambda_le_1)/pi/pi/AR1/AR1)+a0*cos(Lambda_le_1)/pi/AR1)
CL_alpha2=(a0*cos(Lambda_le_2))/(sqrt(1-M*M*cos(Lambda_le_2)*cos(Lambda_le_2)+a0*cos(Lambda_le_2)*a0*cos(Lambda_le_2)/pi/pi/AR2A/AR2A)+a0*cos(Lambda_le_2)/pi/AR2A)
```

```python
display(Math(r'C_{L_\alpha} \Big|_{P_1}=' + '{0:.3}'.format(CL_alpha1) +r'\text{rad}{-1}'))
display(Math(r'C_{L_\alpha} \Big|_{P_{2}}='  + '{0:.3}'.format(CL_alpha2) +r'\text{rad}{-1}'))
```

```python
xac_cr=(XacCr*S1*CL_alpha1+XacCr2*S_ref2A*CL_alpha2)/(S1*CL_alpha2+S_ref2A*CL_alpha1)
xac=xac_cr*c_r
xacmedio=(xac-x_lec)/c_mean
```
